```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('solarpowergeneration.csv')
```

```python
print("Dataset Overview:")
print(df.head())
```

```python
print("Dataset Overview:")
print(df.tail())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['power-generated'], kde=True, color='blue')
plt.title('Distribution of Power Generated')
plt.xlabel('Power Generated (Joules)')
plt.ylabel('Frequency')
plt.show()
```

```python
numerical_features = ['temperature', 'wind-speed', 'visibility', 'humidity', 'average-wind-speed-(period)']
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_features, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_features, 1):
    plt.subplot(2, 3, i)
    sns.scatterplot(x=df[col], y=df['power-generated'])
    plt.title(f'{col} vs Power Generated')
    plt.xlabel(col)
    plt.ylabel('Power Generated (Joules)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x=df['sky-cover'], palette='viridis')
plt.title('Sky Cover Distribution')
plt.xlabel('Sky Cover (0 to 4)')
plt.ylabel('Count')
plt.show()
```
